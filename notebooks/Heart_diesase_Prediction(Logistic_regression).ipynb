## STEP - 3 : FEATURE ENGINEERING

```python
# Label encoding 
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
# Apply label encoding to each categorical column
dataset['Heart Disease'] = le.fit_transform(dataset['Heart Disease'])
```

```python
# 1st 5 lines of the encoded data
dataset.head()
```

## STEP - 4  : DIVIDING THE DATASET

```python
# dividing into  x(input) and y(output)
X = dataset.drop(['Heart Disease', 'ST depression', 'Slope of ST'], axis = 1)
y = dataset['Heart Disease']
# input
X
```

```python
# output
y
```

```python
# shape of x and y
X.shape, y.shape
```

## STEP 5  : SPLLITING AND TRAINING THE MODEL

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

# train data
X_train
```

```python
# test data
X_test
```

```python
# Training the Logistic Regression model on the Training set
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(C=1.0)

classifier = classifier.fit(X_train, y_train)
classifier
```

## STEP  - 6 : PREDICTING THE Y-TEST DATA 

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred
```

```python
# predicted probabilty 
y_pred_proba = classifier.predict_proba(X_test)[:, 1]
y_pred_proba
```

```python
# original and predicted values - concatenation of arry
calculation = pd.DataFrame(np.c_[y_test,y_pred], columns = ["Original Heart Disease","Predict Heart Disease"])
calculation
```

## STEP - 7 : VISUVALIZING THE TRAIN AND TEST DATA

```python
# Visualize Only Train data
plt.plot(X_train,y_train,'r.',label='train_data')
plt.xlabel('X_train')
plt.xlabel('y_train')
plt.legend()
plt.show()
```

```python
# Visualize Only Test data
plt.plot(X_test,y_test,'g.',label='test_data')
plt.xlabel('X_test')
plt.xlabel('y_test')
plt.legend()
plt.show()
```

## STEP 8 : FINDING ACCURACY OF TRAINING AND TESTING DATA

```python
# accuracy
print("Training Accuracy :", classifier.score(X_train, y_train))
print("Testing Accuracy :", classifier.score(X_test, y_test))
```

```python
# importing metrics
from sklearn.metrics import accuracy_score
```

```python
# accuracy score
accuracy_score = accuracy_score(y_test,y_pred)
```

```python
print(f"Accuracy score of testing data: {accuracy_score}")
```

## STEP - 9 : CLASSIFICATION MODELS 

```python
# importing classification and confusion matrix
from sklearn.metrics import confusion_matrix, auc
```

```python
# confusion matrix
from sklearn.metrics import ConfusionMatrixDisplay

# Compute confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Not Class 0', 'Class 0'])
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
```

```python
# ROC curve
from sklearn.metrics import roc_curve, roc_auc_score

# Compute ROC curve
fpr, tpr, _ = roc_curve(y_test, y_pred)

# Compute ROC AUC
roc_auc = roc_auc_score(y_test, y_pred_proba)

# Plot ROC curve
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
```

```python
# importing precision recall curve
from sklearn.metrics import precision_recall_curve

# Compute Precision-Recall curve
precision, recall, threshold = precision_recall_curve(y_test, y_pred_proba)

# Plot Precision-Recall curve
plt.figure()
plt.plot(recall, precision, color='blue', lw=5)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.show()
```

```python
# overall performance of the classification models
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
